# Importing various datasets which are present in Sklearn.datasets

### IRIS DATASET

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X,y = iris.data , iris.target
```

```python
X[:]
```

```python
y[:]
```

```python
len(X)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred_rfc = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report , mean_absolute_error , mean_squared_error
```

```python
print(classification_report(y_test,y_pred_rfc))
```

### Fetch California Housing

```python
from sklearn.datasets import fetch_california_housing
```

```python
fch = fetch_california_housing()
```

```python
X , y = fch.data , fch.target
```

```python
len(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X[:6]
#y[:6]
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor , RandomForestRegressor
```

```python
hgbr = HistGradientBoostingRegressor( max_leaf_nodes=15, random_state=101, early_stopping=False)
```

```python
hgbr.fit(X_train,y_train)
```

```python
y_pred_hgbr = hgbr.predict(X_test)
```

```python
#print(classification_report(y_test,y_pred_hgbr))
```

```python
y_pred_hgbr.round()
```

```python
y_test
```

```python
mae=mean_absolute_error(y_test,y_pred_hgbr)
mae
```

```python
mse=np.sqrt(mean_squared_error(y_test,y_pred_hgbr))
mse
```

```python
y_test.mean()
```

```python
100*mae / y_test.mean()
```

```python
import joblib
N_CORES = joblib.cpu_count(only_physical_cores=True)
N_CORES
```

```python
rfr = RandomForestRegressor(min_samples_leaf=5, random_state=101, n_jobs=N_CORES)
```

```python
rfr.fit(X_train,y_train)
```

```python
y_pred_rfr = rfr.predict(X_test)
```

```python
mae=mean_absolute_error(y_test,y_pred_rfr)
```

```python
mae
```

```python
mse=np.sqrt(mean_squared_error(y_test,y_pred_rfr))
mse
```

```python
100*mae / y_test.mean()
```

### Dump SVMlight File

```python
output_file = 'my_datasets.dump_svmlight_file'
```

```python
dump_svmlight_file(X,y,output_file)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X[0]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
mean_absolute_error(y_test,y_pred_lr)
```
